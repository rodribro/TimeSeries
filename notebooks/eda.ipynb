#### <b> One Hot Encoding </b>

* In order to be able to use categorical data to perform EDA we need to transform categorical data into a numerical format, hence we'll encode the values of the columns of interest.

```python
biometric_df['CODE'].nunique(), biometric_df['CODE'].unique()
```

```python
biometric_df['Sample'].nunique(), biometric_df['Sample'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_code = LabelEncoder()
```

<b> Encoding 'CODE' column </b>

```python
le_code.fit(biometric_df['CODE'])
```

```python
list(le_code.classes_)
```

```python
code_data = le_code.transform(biometric_df['CODE'])
```

```python
code_data
```

```python
biometric_df['CODE'] = code_data
```

```python
biometric_df
```

<b> Encoding 'Sample' column </b>

```python
le_sample = LabelEncoder()
```

```python
le_sample.fit(biometric_df['Sample'])
sample_data = le_sample.transform(biometric_df['Sample'])
```

```python
biometric_df['Sample'] = sample_data
```

```python
biometric_df
```

#### <b> Plotting the environmental variables for a given plant </b>

```python
#Function that allows you to plot the values of the Envionmental Variables for a given plant
def environmentVariables(df, plant):
    
    plt.figure(figsize=(24,12))
    
    temp = df[df["Number"]==plant]["Mean. Temp."]
    hum = df[df["Number"]==plant]["Mean. Hum."]
    hgt = df[df["Number"]==plant]["Height"]
    #ph = df[df["Number"]==plant]["pH Level"]

    plt.suptitle(f"Environmental Variables for plant {plant}:", fontsize=15, y=1.0)
    
    plt.subplot(131)
    plt.plot(temp, label="Temperature (C)", color='r')
    plt.title("Mean Temperature (C)")
    plt.xticks(rotation=45)

    plt.subplot(132)
    plt.plot(hum, label="Humidity (%)", color='b')
    plt.title("Mean Humidity")
    plt.xticks(rotation=45)

    plt.subplot(133)
    plt.plot(hgt, label="Plant Height", color='y')
    plt.title("Plant Height")
    plt.xticks(rotation=45)
    
    plt.tight_layout()
```

```python
environmentVariables(biometric_df, 45)
```
