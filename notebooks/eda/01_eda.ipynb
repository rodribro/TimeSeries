#### <b> Plotting the environmental variables for a given plant </b>

```python
#Function that allows you to plot the values of the Envionmental Variables for a given plant
def environmentVariables(df, plant):
    
    plt.figure(figsize=(24,12))
    
    temp = df[df["Number"]==plant]["Mean. Temp."]
    hum = df[df["Number"]==plant]["Mean. Hum."]
    hgt = df[df["Number"]==plant]["Height"]
    #ph = df[df["Number"]==plant]["pH Level"]

    plt.suptitle(f"Environmental Variables for plant {plant}:", fontsize=15, y=1.0)
    
    plt.subplot(131)
    plt.plot(temp, label="Temperature (C)", color='r')
    plt.title("Mean Temperature (C)")
    plt.xticks(rotation=45)

    plt.subplot(132)
    plt.plot(hum, label="Humidity (%)", color='b')
    plt.title("Mean Humidity")
    plt.xticks(rotation=45)

    plt.subplot(133)
    plt.plot(hgt, label="Plant Height", color='y')
    plt.title("Plant Height")
    plt.xticks(rotation=45)
    
    plt.tight_layout()
```

##### <b> PCA </b>

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
pca.fit(scaled_data) # calculate loading scores and variation for each principal component
```

```python
pca_data = pca.transform(scaled_data) # generate coordinates for a PCA graph based on the loading scores and scaled data
```

```python
# Principal Component loadings (coeffiecients)
loadings = pca.components_.T
```

```python
import matplotlib.pyplot as plt

# PCA is already fitted
explained_variance = pca.explained_variance_ratio_

# Define a colormap for unique colors for each arrow
colors = plt.cm.get_cmap('tab10', len(enc_df.columns))

# Plotting the loading plot with feature names as a legend
plt.figure(figsize=(10, 8))

# Store color handles for the legend
handles = []

# Plot arrows and label them with numbers corresponding to features
for i, feature in enumerate(enc_df.columns):
    color = colors(i)
    arrow = plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], color=color, alpha=0.75)
    handles.append(arrow)
    plt.text(loadings[i, 0] * 1.15, loadings[i, 1] * 1.15, str(i + 1), color=color, ha='center', va='center')

# Adding labels for the axes with variance percentages
explained_variance = pca.explained_variance_ratio_
plt.xlabel(f'Principal Component 1 ({explained_variance[0]:.2%} variance)')
plt.ylabel(f'Principal Component 2 ({explained_variance[1]:.2%} variance)')
plt.title('Loading Plot for PCA')

# Draw origin lines
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Create a legend mapping feature names to the numbered labels
legend_labels = [f"{i + 1}: {feature}" for i, feature in enumerate(enc_df.columns)]
plt.legend(handles, legend_labels, loc='upper right', bbox_to_anchor=(1.3, 1.05))

plt.grid()
plt.show()
```

<b> Feature Grouping </b>

1. Cross-sectional Size Group:

* Combines Diameter, Perpendicular, and Average Leaf Thickness
    * These load together strongly on PC1 (42.08% variance)
    * Creates a single metric for plant girth/width


2. Temperature Metrics:

* Keeps temperature range and mean as separate features
    * Captures both absolute temperature and variability
    * Important due to strong loadings on both PC1 and PC2


3. Height (kept separate):

* Maintains height as an independent feature
    * Justified by its distinct loading pattern on PC2
    * Represents a different aspect of growth than cross-sectional measurements


4. Humidity Metrics:

* Groups basic humidity metrics (max, min, mean) into one stability measure
    * Keeps humidity variation separate due to its distinct pattern
    * Reflects the importance of both absolute humidity and its variability


5. Temporal Features (optional):

* Combined into a single time metric
    * Can be dropped entirely given weak loadings
    * Included as optional since temporal patterns might still be of interest

```python
# Calculate the percent of variation each principal coponent accounts for
per_var = np.round(pca.explained_variance_ratio_*100, decimals=1)
```

```python
# Labels for each principal component
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
```

```python
# Scree plot
plt.bar(x=range(1, len(per_var)+1), height = per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.xticks(rotation=60)
plt.title('Scree Plot')
plt.show()
```
