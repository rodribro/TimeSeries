```python
#IMPORT PACKAGES
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#IMPORT DATASET
df = pd.read_excel(r'data_lr_white.xlsx', header=1)
print(df)
```

```python
# Filter rows where Variety is 'TA'

ta = df[df['Variety'] == 'TA']
#ta = ta[ta['Date'] == 1]

tb = df[df['Variety'] == 'TB']
#tb = tb[tb['Date'] == 1]

print(ta)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming ta is your DataFrame containing the data for 'TA'

# Extracting features from the DataFrame
features = ta.iloc[:, 3:]  # Assuming your features start from the second column


# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform PCA
pca = PCA()
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame for the principal components
columns = [f"PC{i+1}" for i in range(principal_components.shape[1])]
principal_df = pd.DataFrame(data=principal_components, columns=columns)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

print("Explained Variance Ratio:")
for i, ratio in enumerate(explained_variance_ratio):
    print(f"PC{i+1}: {ratio:.2f}")

# Optionally, you can also visualize the principal components or perform further analysis
```

```python
import matplotlib.pyplot as plt

# Plotting the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(principal_df['PC1'], principal_df['PC2'], alpha=0.8)
plt.title('PCA Scatter Plot')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1', fontsize=20)
plt.ylabel('Principal Component - 2', fontsize=20)
plt.title("Principal Component Analysis of Lettuce Nitrogen Content", fontsize=20)

targets = ['zero', 'two', 'four', 'eight'] 
colors = ['r', 'g', 'b', 'yellow']

for target, color in zip(targets, colors):
    indicesToKeep = df['Status'] == target
    plt.scatter(principal_df.loc[indicesToKeep, 'PC1'], principal_df.loc[indicesToKeep, 'PC2'], c=color, s=50, label=target, alpha=0.8)

plt.legend(prop={'size': 15})
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Assuming 'ta' is your DataFrame containing the data for 'TA'

# Extracting features (X) and target (y) from the DataFrame
X = ta.iloc[:, 3:]  # Assuming your features start from the fifth column
y = ta['Status']    # Assuming 'Status' is the target column

# Initialize and fit the LDA model
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)

# Transform the features to the LDA space
X_lda = lda.transform(X)

# Print the explained variance ratio
print("Explained Variance Ratio:")
print(lda.explained_variance_ratio_)

# Optionally, you can perform further analysis or visualization with the transformed features X_lda
```

```python
import matplotlib.pyplot as plt

# Define colors for each category
category_colors = {'zero': 'red', 'two': 'blue', 'four': 'green', 'eight': 'orange'}  # Adjust colors as needed

# Plotting the LDA components with colors according to categories
plt.figure(figsize=(8, 6))
for category, color in category_colors.items():
    X_category = X_lda[y == category]
    plt.scatter(X_category[:, 0], X_category[:, 1], color=color, label=category, alpha=0.8)

plt.title('Linear Discriminant Analysis of Lettuce Nitrogen Content - TA variety')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
