# MSC preprocessing

```python
def msc(input_data, reference=None):
    """
        :msc: Scatter Correction technique performed with mean of the sample data as the reference.
        :param input_data: Array of spectral data
        :type input_data: DataFrame
        :returns: data_msc (ndarray): Scatter corrected spectra data
    """
    eps = np.finfo(np.float32).eps
    input_data = np.array(input_data, dtype=np.float64)
    ref = []
    sampleCount = int(len(input_data))

    # mean centre correction
    for i in range(input_data.shape[0]):
        input_data[i,:] -= input_data[i,:].mean()
    
    # Get the reference spectrum. If not given, estimate it from the mean
    # Define a new array and populate it with the corrected data    
    data_msc = np.zeros_like(input_data)
    for i in range(input_data.shape[0]):
        for j in range(0, sampleCount, 10):
            ref.append(np.mean(input_data[j:j+10], axis=0))
            # Run regression
            fit = np.polyfit(ref[i], input_data[i,:], 1, full=True)
            # Apply correction
            data_msc[i,:] = (input_data[i,:] - fit[0][1]) / fit[0][0]
    
    return (data_msc)
```

```python

# Get the weights of each variable for each discriminant
weights = lda.coef_

# Determine the threshold for each discriminant based on the distribution of the weights
n_ld = weights.shape[0]
thresholds = []
for i in range(n_ld):
    w = weights[i]
    # Compute the interquartile range (IQR) of the weights
    q1, q3 = np.percentile(w, [25, 75])
    iqr = q3 - q1
    # Set the threshold to be 1.5 times the IQR above the upper quartile
    threshold = q3 + 1.5*iqr
    thresholds.append(threshold)

# Determine the number of important variables for each discriminant based on the thresholds
n_vars = weights.shape[1]
n_important_vars = [sum(w > threshold) for w, threshold in zip(weights, thresholds)]

# Get the names of the important variables for each discriminant
important_vars = []
for i in range(n_ld):
    w = weights[i]
    mask = w > thresholds[i]
    var_names = df1.columns[mask]
    important_vars.append(', '.join(var_names))

# Create a DataFrame with the results
df_results = pd.DataFrame({'Discriminant': range(1, n_ld+1), 'N_important_vars': n_important_vars, 'Important_vars': important_vars})
display(df_results)

# Save the DataFrame to a CSV file
df_results.to_csv('lda_results.csv', index=False)

# Plot the distribution of the weights for each discriminant
fig, axs = plt.subplots(nrows=n_ld, figsize=(8, 6))
for i in range(n_ld):
    axs[i].hist(weights[i], bins=20)
    axs[i].axvline(thresholds[i], color='red', linestyle='--')
    axs[i].set_title(f"Discriminant {i+1}")
fig.tight_layout()
plt.show()

# Print the number of important variables for each discriminant
for i in range(n_ld):
    print(f"Discriminant {i+1}: {n_important_vars[i]} important variables")

```

In this example, the code fits an LDA model to the iris dataset and obtains the weights of each variable for each discriminant using the coef_ attribute. Instead of using a fixed threshold for all discriminants, the code determines the threshold for each discriminant based on the distribution of the weights. Specifically, it computes the interquartile range (IQR) of the weights and sets the threshold to be 1.5 times the IQR above the upper quartile. This ensures that the threshold is sensitive to the distribution of the weights and can capture outliers and extreme values.

The code then determines the number of important variables for each discriminant based on the thresholds and prints the results. Finally, it plots the distribution of the weights for each discriminant and shows the threshold as a red dashed line.

Note that the choice of the method for setting the threshold may depend on the distribution and characteristics of the weights, and you may need to experiment with different methods to find the most appropriate one for your data. Also, keep in mind that the interpretation of the weights may depend on the scaling and normalization of the variables, and you may need to perform additional preprocessing steps before interpreting the results.
